# Exemplos de Técnicas de Prompt Engineering com Google Gemini API (Código Corrigido)

```python

import os
from dotenv import load_dotenv

# Carregar variáveis de ambiente do arquivo .env
load_dotenv()

import google.generativeai as genai

# Configuração da chave da API
genai.configure(api_key=os.environ["GEMINI_API_KEY"])

# Configuração do modelo base
generation_config = {
  "temperature": 1,
  "top_p": 0.95,
  "top_k": 40,
  "max_output_tokens": 8192,
  "response_mime_type": "text/plain",
}
```

## 1. Zero-shot Prompting

Zero-shot prompting involves providing a model with a task or question without any prior examples or context. The model must rely solely on its pre-trained knowledge to generate a response. This technique is useful for quickly obtaining answers or generating content without the need for extensive setup or training data.

In the context of the Google Gemini API, zero-shot prompting can be used to ask the model questions or request information directly, leveraging the model's extensive training on diverse datasets to produce relevant and accurate responses.

### Example

In the following example, we use zero-shot prompting to ask the model about the impact of artificial intelligence on education:



```python

# Zero-shot Prompting
model = genai.GenerativeModel(
  model_name="gemini-1.5-flash-8b",
  generation_config=generation_config,
  system_instruction="Answer questions as accurately as possible without additional context."
)

chat_session = model.start_chat(history=[])
response = chat_session.send_message("What is the impact of artificial intelligence on education?")
print(response.text)
```

## 2. Few-shot Prompting

Few-shot prompting involves providing a model with a few examples of the desired output before asking it to generate a response. This technique helps the model understand the format and context of the task, leading to more accurate and relevant outputs. Few-shot prompting is particularly useful when the task requires specific formatting or when the model needs to follow a particular pattern.

### Example

In the following example, we use few-shot prompting to ask the model about the applications of AI in different sectors:


```python
# Few-shot Prompting
model = genai.GenerativeModel(
  model_name="gemini-1.5-flash-8b",
  generation_config=generation_config,
  system_instruction="Respond concisely and follow the pattern of the provided examples."
)

chat_session = model.start_chat(
    history=[
        {
            "role": "user",
            "parts": [
                "How does machine learning improve healthcare?"
            ]
        },
        {
            "role": "model",
            "parts": [
                "Machine learning improves healthcare by providing predictive analytics, personalizing treatment, and improving diagnostic accuracy."
            ]
        },
        {
            "role": "user",
            "parts": [
                "What are some applications of AI in finance?"
            ]
        },
        {
            "role": "model",
            "parts": [
                "AI in finance is used for fraud detection, personalized financial advice, algorithmic trading, and risk management."
            ]
        }
    ]
)

response = chat_session.send_message("How is AI used in retail?")
print(response.text)
```

## 3. Chain-of-Thought Prompting

Chain-of-Thought Prompting involves guiding the model to provide answers step by step, explaining each part of the reasoning process. This technique helps in generating more detailed and logical responses by breaking down complex questions into smaller, manageable parts.

### Example

In the following example, we use Chain-of-Thought Prompting to ask the model about how AI can help in the process of drug discovery:

```python

# Chain-of-Thought Prompting
model = genai.GenerativeModel(
  model_name="gemini-1.5-flash-8b",
  generation_config=generation_config,
  system_instruction="Provide answers step by step, explaining each part of the reasoning."
)

chat_session = model.start_chat(history=[])
response = chat_session.send_message("How can AI help in the process of drug discovery?")
print(response.text)
```

## 4. Self-Consistency

Self-Consistency involves generating multiple responses to the same prompt and selecting the most consistent or frequent answer. This technique helps improve the reliability and accuracy of the generated content by leveraging the consensus among multiple outputs.

### Example

In the following example, we use Self-Consistency to determine the benefits of renewable energy:


```python

# Self-Consistency
model = genai.GenerativeModel(
  model_name="gemini-1.5-flash-8b",
  generation_config=generation_config,
  system_instruction="Generate consistent and well-supported responses to enhance accuracy."
)

chat_session = model.start_chat(history=[])
responses = [chat_session.send_message("What are the benefits of renewable energy?").text for _ in range(5)]
final_answer = max(set(responses), key=responses.count)  # Seleciona a resposta mais frequente
print(final_answer)
```

## 5. Prompt Chaining

Prompt Chaining involves linking multiple prompts together to build upon previous responses, creating a more complex and detailed output. This technique allows for the development of nuanced and comprehensive answers by iteratively refining the context and incorporating new information.

### Example

In the following example, we use Prompt Chaining to explore the main factors in AI ethics, the challenges in AI regulation, and how organizations can address these challenges:

1. **Initial Prompt:** "What are the main factors in AI ethics?"
2. **Follow-up Prompt:** "Given the factors '[response1]', what are the challenges in AI regulation?"
3. **Final Prompt:** "Considering '[response2]', how can organizations address these challenges?"

By chaining these prompts, we can develop a detailed and interconnected understanding of AI ethics, regulatory challenges, and potential solutions.

```python

# Prompt Chaining
model = genai.GenerativeModel(
  model_name="gemini-1.5-flash-8b",
  generation_config=generation_config,
  system_instruction="Provide detailed responses that build upon prior answers to develop complex ideas."
)

chat_session = model.start_chat(history=[])
response1 = chat_session.send_message("What are the main factors in AI ethics?")
response2 = chat_session.send_message(f"Given the factors '{response1.text}', what are the challenges in AI regulation?")
response3 = chat_session.send_message(f"Considering '{response2.text}', how can organizations address these challenges?")
print(response3.text)
```

## 7. Retrieval Augmented Generation (RAG) geting context from PDF file

Retrieval Augmented Generation (RAG) is a technique that combines the power of retrieval-based methods with generative models to provide more accurate and contextually relevant responses. This approach involves retrieving relevant documents or pieces of information from a large corpus and using them as context for the generative model to produce a comprehensive answer.

In this example, we use RAG to answer questions about the Tupi-Guarani dictionary. The process involves the following steps:

1. **Extract Text from PDF:**
    We extract text from a PDF document containing the Tupi-Guarani dictionary using the `fitz` library (PyMuPDF).

2. **Split Text into Smaller Documents:**
    The extracted text is split into smaller, manageable documents using the `RecursiveCharacterTextSplitter` from `langchain_text_splitters`.

3. **Create Embeddings:**
    We create embeddings for the smaller documents using the `SentenceTransformer` model from Hugging Face.

4. **Configure FAISS Index:**
    The generated embeddings are added to a FAISS index for efficient similarity search.

5. **Retrieve Relevant Documents:**
    When a question is asked, we retrieve the most relevant documents from the FAISS index based on the similarity of their embeddings to the question's embedding.

6. **Generate Answer:**
    The retrieved documents are used as context for the generative model (Google Gemini API) to generate a comprehensive answer.

### Example

In the following example, we ask the model about the Tupi-Guarani term "Ababá":

1. **Question:** "What is Ababá?"
2. **Retrieved Context:** Relevant documents from the Tupi-Guarani dictionary.
3. **Generated Answer:** "Ababá was a Tupi-Guarani indigenous tribe that lived near the headwaters of the Corumbiara River in Mato Grosso, Brazil."

By integrating retrieval-based methods with generative models, RAG provides detailed and contextually accurate responses, leveraging both the knowledge in the documents and the generative capabilities of the model.

```python
# Import Necessary Libraries
import fitz  # PyMuPDF for PDF manipulation
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
from langchain_core.documents import Document

# Load the embedding model from Hugging Face
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')  # or another available model

# Function to extract text from PDF
def extract_text_from_pdf(pdf_path):
    text = ""
    with fitz.open(pdf_path) as pdf_document:
        for page in pdf_document:
            text += page.get_text()
    return text

# Extract the content of the PDF
pdf_path = "kb/dicitupi.pdf"
pdf_content = extract_text_from_pdf(pdf_path)

# Split the text into smaller documents for better processing
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter()
pdf_documents = text_splitter.split_text(pdf_content)

# Create embeddings for the documents using the Hugging Face model
embeddings = [embedding_model.encode(text) for text in pdf_documents]

# Configure FAISS with the generated embeddings
dimension = embeddings[0].shape[0]  # dimension of the embeddings
index = faiss.IndexFlatL2(dimension)  # FAISS index for L2 search
index.add(np.array(embeddings))  # add the embeddings to the index

# Function to find relevant documents
def retrieve_documents(question, k=5):
    question_embedding = embedding_model.encode(question).reshape(1, -1)
    distances, indices = index.search(question_embedding, k)  # find k closest documents
    relevant_documents = [pdf_documents[i] for i in indices[0]]
    return relevant_documents

# Function to ask questions using the Gemini model with the extracted context from the PDF

generation_config = {
    "temperature": 0.7
}

# Define the model with the new configuration
model = genai.GenerativeModel(
    model_name="gemini-1.5-flash-8b",
    generation_config=generation_config,
    system_instruction="Integrate provided information with additional knowledge to create a comprehensive answer."
)

# Function to extract context and answer the question
def ask_pdf(question):
    # Retrieve relevant documents
    relevant_documents = retrieve_documents(question)
    context = " ".join(relevant_documents)

    # Send the question to the model with the context
    chat_session = model.start_chat(history=[])
    response = chat_session.send_message(f"Using the information: '{context}', answer the question: '{question}'")
    
    return response.text

# Example usage: ask about the content of the PDF
question = "What is Ababá?"
answer = ask_pdf(question)
print(answer)
```
